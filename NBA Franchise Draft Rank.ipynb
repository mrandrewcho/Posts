```python
import pandas as pd
from pandas import Series,DataFrame,read_html
import numpy as np

from bs4 import BeautifulSoup
import html5lib

import sys
reload(sys)
sys.setdefaultencoding('utf-8')
```

```python
# start by creating a url template to use in the loop
url_template = "http://www.basketball-reference.com/draft/NBA_{year}.html"
```

```python
# create an empty DataFrame to append each draft year DataFrame to
draft_dframe = pd.DataFrame()
```

```python
for year in range(1979, 2018): # Will build the DataFrame for each year of interest
    url = url_template.format(year = year) # Grabbing the correct url
    
    dframe_list_year = pd.io.html.read_html(url)
    dframe_year = dframe_list_year[0]
    
    # Dropping the NaN filled columns
    dframe_year.drop(dframe_year.columns[[0]],inplace=True,axis=1)
    
    # Add in a column for the draft year
    dframe_year.insert(0, 'Draft_Yr', year)
    
    # Append to the big DataFrame
    draft_dframe = draft_dframe.append(dframe_year, ignore_index=True)
```

```python
pd.set_option('display.max_columns', None)
column_names = ['Draft_Yr','Pk','Team','Player','College','Yrs','Games','Minutes Played','PTS','TRB','AST','FG_Percentage','3P_Percentage','FT_Percentage','Minutes per Game','Points per Game','TRB per game','Assits per Game','Win Share','WS_per_game','BPM','VORP']
draft_dframe.columns = column_names
```

```python
draft_dframe
```

```python
# Dropping the rows that served as breaks for different rounds of the draft
draft_dframe = draft_dframe[draft_dframe.Pk.notnull()]
draft_dframe = draft_dframe[draft_dframe['Pk'] != 'Pk']
```

```python
draft_dframe
```

```python
# changing the remaining NaN's to zeroes
draft_dframe = draft_dframe.fillna(0)
# reindexing to align the row index correctly
draft_dframe.index = range(3560)
```

```python
draft_dframe
```

```python
draft_dframe['Team'].unique()
```

```python
draft_dframe = draft_dframe.replace('NJN', 'BRK')
draft_dframe = draft_dframe.replace('CHA', 'CHH')
draft_dframe = draft_dframe.replace('CHO', 'CHH')
draft_dframe = draft_dframe.replace('VAN', 'MEM')
draft_dframe = draft_dframe.replace('NOH', 'NOP')
draft_dframe = draft_dframe.replace('NOK', 'NOP')
draft_dframe = draft_dframe.replace('SEA', 'OKC')
draft_dframe = draft_dframe.replace('WSB', 'WAS')
draft_dframe = draft_dframe.replace('KCK', 'SAC')
draft_dframe = draft_dframe.replace('SDC', 'LAC')
draft_dframe['Team'].unique()
```

```python
draft_dframe
```

```python
# Converting the data types to the proper type
numeric_columns = column_names
del numeric_columns[1:4] # Dropping the string columns 'Player' 'Team' 'College' from this list

# Converting data types using the numeric_columns list
for column in numeric_columns:
    draft_dframe[column] = pd.to_numeric(draft_dframe[column], errors='coerce')
draft_dframe['Pk'] = draft_dframe['Pk'].str.replace(',','').astype(int)
draft_dframe.dtypes
```

```python
avg_win_shares_at_pick = draft_dframe.groupby(['Pk'])['Win Share'].mean()
```

```python
draft_dframe['Average Win Share at Pick'] = ''
```

```python
draft_dframe = draft_dframe[draft_dframe['Draft_Yr'] < 2013]
draft_dframe = draft_dframe[draft_dframe['Pk'] < 61]
```

```python
for pick in draft_dframe['Pk']:
    draft_dframe['Average Win Share at Pick'][draft_dframe['Pk'] == pick] = avg_win_shares_at_pick[pick]
```

```python
draft_dframe['Draft Score (Win Share - Average Win Share at Pick)'] = draft_dframe['Win Share'] - draft_dframe['Average Win Share at Pick']
```

```python
draft_dframe
```

```python
draft_dframe.sort_values("Draft Score (Win Share - Average Win Share at Pick)", ascending = False)
```

```python
franchise_draft_scores = draft_dframe.groupby(['Team'])['Draft Score (Win Share - Average Win Share at Pick)'].sum()
```

```python
franchise_draft_scores.sort_values(ascending = False)
```
