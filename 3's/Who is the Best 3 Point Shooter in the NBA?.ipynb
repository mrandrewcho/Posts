```python
#function to calculate the effective 3 point percentage
def da3FGpct_calc(inputs):
    ThreePM = inputs['3PM']
    
    Short_3PM = inputs['20-24 FG3M']
    MediumToLong_3PM = inputs['25-34 FG3M']
    MediumToLong_multiplier = inputs['20-24 Average']/inputs['25-34 Average']
    ShotDistance_3PA = inputs['20-24 FG3A'] + inputs['25-34 FG3A']
    
    CatchAndShoot_3PM = inputs['Catch and Shoot FG3M']
    PullUp_3PM = inputs['Pull Ups FG3M']
    PullUp_multiplier = inputs['Catch and Shoot Average']/inputs['Pull Ups Average']
    ShotType_3PA = inputs['Catch and Shoot FG3A'] + inputs ['Pull Ups FG3A']
    
    Contested_3PM = inputs['Contested FG3M']
    Uncontested_3PM = inputs['Uncontested FG3M']
    Contested_multiplier = inputs['Uncontested Average']/inputs['Contested Average']
    Open_3PM = inputs['Open FG3M']
    DefenderDistance_3PA = inputs['Contested FG3A'] + inputs['Uncontested FG3A']
    
    pct = ((1.0/3) * ((Short_3PM + (MediumToLong_3PM * MediumToLong_multiplier)) / ShotDistance_3PA)) + ((1.0/3) * ((CatchAndShoot_3PM + (PullUp_3PM * PullUp_multiplier)) / ShotType_3PA)) + ((1.0/3) * ((Uncontested_3PM + (Contested_3PM * Contested_multiplier)) / DefenderDistance_3PA))  
    return pct
```

```python
merged_df['Difficulty Accounted 3 Point Percentage'] = merged_df.iloc[:, 1:].sum(axis=1)
for index, row in merged_df.iterrows():
    row['Difficulty Accounted 3 Point Percentage'] = da3FGpct_calc(row)
    merged_df.loc[row.name,'Difficulty Accounted 3 Point Percentage'] = da3FGpct_calc(row)
```
